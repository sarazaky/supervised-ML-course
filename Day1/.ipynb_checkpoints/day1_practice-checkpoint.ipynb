```python
import math
import numpy as np
import matplotlib.pyplot as plt

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
x = 2 * math.pi * np.random.rand(25, 1)

x = np.sort(x, axis=0)

y = np.sin(x).ravel()

t = y + 0.01 - np.random.rand(25)
plt.plot(x, t, "*")
```

```python
x_poly = PolynomialFeatures(degree=4).fit_transform(x)


reg = LinearRegression().fit(x_poly, t)

y_predict = reg.predict(x_poly)

plt.plot(x, t, "*")
plt.plot(x, y_predict)
```

```python
reg.score(x_poly, t)
```
