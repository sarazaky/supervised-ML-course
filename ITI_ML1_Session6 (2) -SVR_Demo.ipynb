```python
import numpy as np
from sklearn.svm import SVR
import matplotlib.pyplot as plt
```

Dataset Preparation

```python
#genarate 50 points of data between 0 and 5
x= 5* np.random.rand(50, 1)
# sort the numbers
x= np.sort(x, axis =0)
#Assume a true function y = sin (x)
y=np.sin(x).ravel()
plt.plot(x,y,'*')
```

```python
#add noise to a subset of y values
y[::5] += 3*(0.5-np.random.rand(10))
plt.plot(x,y,'*')
```

Using Linear SVR

```python
# Fit regression model
svr_linear = SVR(kernel='linear', C=100, gamma='auto', epsilon=.1)
# For demonstration, let's fit on the data and use the model to predict 
#the values of y given x 
svr_linear.fit(x, y)
y_predict = svr_linear.predict(x)
plt.plot(x,y,'o')
plt.plot(x,y_predict)
plt.legend(["True data", "Predicted data"])
```

Using RBF Kernels

```python
# Fit regression model
svr_rbf = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.1)
# For demonstration, let's fit on the data and use the model to predict 
#the values of y given x 
svr_rbf.fit(x, y)
y_predict = svr_rbf.predict(x)
plt.plot(x,y,'o')
plt.plot(x,y_predict)
plt.legend(["True data", "Predicted data"])
```

Use Poly Kernel

```python
# Fit regression model
svr_poly = SVR(kernel='poly', C=100, gamma='auto', degree=3, epsilon=.1, coef0=1)
# For demonstration, let's fit on the data and use the model to predict 
#the values of y given x 
svr_poly.fit(x, y)
y_predict = svr_poly.predict(x)
plt.plot(x,y,'o')
plt.plot(x,y_predict)
plt.legend(["True data", "Predicted data"])
```

Use different epsilon

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 10))
fig.suptitle('Impact of Decreasing Epsilon')
svr_E_1 = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=.01)
svr_E_2 = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.1)
svr_E_3 = SVR(kernel='rbf', C=100, gamma=0.1, epsilon=0.3)
ax1.plot(x, svr_E_1.fit(x, y).predict(x))
ax1.plot(x,y,'o')
ax1.set_title('epsilon=.01')
ax2.plot(x, svr_E_2.fit(x, y).predict(x))
ax2.plot(x,y,'o')
ax2.set_title('epsilon=10')
ax3.plot(x, svr_E_3.fit(x, y).predict(x))
ax3.plot(x,y,'o')
ax3.set_title('epsilon=100')
```
